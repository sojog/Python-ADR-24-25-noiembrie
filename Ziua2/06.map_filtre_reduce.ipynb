# Python suporta programare functionala

# 1. MAP - mapeaza toate elemente folosind o functie

```python
## Mapara a elementelor
lista =  [11, 22, 33, 44, 55]

def transforma(x):
    return x * 10
```

```python
# Clasic
o_noua_lista = []
for i in lista:
    o_noua_lista.append(transforma(i))
o_noua_lista
```

```python
# Folosind functia Map
list(map(transforma, lista))
```

```python
lista =  [11, 22, 33, 44, 55]
list(map(str, lista))
```

```python
lista =  [11, 22, 33, 44, 55]
list(map(float, lista))
```

# 2. Filtrare - filtreaza toate elemente dupa o regula

```python
## Mapara a elementelor
lista =  [11, 22, 33, 44, 55]

def filtreaza_pare(x):
    return x % 2 == 0     # returneaza True or False
```

```python
# Clasic
o_noua_lista = []
for i in lista:
    if filtreaza_pare(i):
        o_noua_lista.append(i)
o_noua_lista
```

```python
## Filter
lista =  [11, 22, 33, 44, 55]
list(filter(filtreaza_pare ,lista))
```

## 3. Reduce

```python
## Reducerea elementelor la unul singur
lista =  [ 44, 22, 33, 55, 11]

def cel_mai_mare_numar(x, y):
    return x if x > y else y
    
```

```python
lista =  [ 44, 22, 33, 55, 11]
# lista =  [ 44,   33, 55, 11]
# lista =  [ 44, 55, 11]
# lista =  [55, 11]
# lista =  [55
```

```python
from functools import reduce
```

```python
lista =  [ 44, 22, 33, 55, 11]

reduce(cel_mai_mare_numar,lista)
```

```python

def suma_numerelor(x, y):
    return x + y

reduce(suma_numerelor, [ 1, 2, 3, 4, 5])
```
