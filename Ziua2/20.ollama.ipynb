# Pentru a comunica cu ollama

```python
OLLAMA_MODEL = "qwen2.5-coder:14b"
```

```python
standar_prompt = "fa-mi un rezumat la modulul pandas din python"
```

```python
prompt = input(f"Ce vrei sa il intrebi pe. {OLLAMA_MODEL} ")
```

```python
payload ={
    "model" :   OLLAMA_MODEL    ,
    "prompt":     standar_prompt ,
    "stream": False
}


response = requests.post("http://localhost:11434/api/generate", json=payload)
response
```

```python
with open("stream_response.txt", "wb") as file_writer:
    file_writer.write(response.content)
```
