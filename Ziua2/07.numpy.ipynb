# 2 - Filtrare

```python
arr = np.array([11, 22, 33, 44, 55, 66, 77]) 
arr % 2 == 0
```

```python
arr = np.array([11, 22, 33, 44, 55, 66, 77]) 
arr [ arr % 2 == 0 ]
```

```python
arr = np.array([11, 22, 33, 44, 55, 66, 77]) 
conditie = arr % 2 == 0 
arr [ conditie ]
```

```python
arr = np.array([11, 22, 33, 44, 55, 66, 77]) 
conditie_1 = arr % 2 == 0 
arr[conditie_1] 
```

```python
arr = np.array([11, 22, 33, 44, 55, 66, 77]) 
conditie_2 = arr > 30
arr[conditie_2] 
```

```python
arr = np.array([11, 22, 33, 44, 55, 66, 77]) 
conditie_1 = arr % 2 == 0 
conditie_2 = arr > 30
arr[conditie_1 & conditie_2] 
```
