```python
# Creaţi o funcție în care utilizatorul va insera o sumă întreagă.

# Programul trebuie să calculeze şi să afişeze numărul corespunzător de bancnote pentru respectiva sumă, unde în program sunt definite bancnotele cu valorile: 10, 5, 2 şi 1

# De exemplu pentru suma 128, sunt 12 bancnote de 10, una de 5, una de 2 și una de 1.
```

```python
suma = 128
```

```python
BANCNOTE = [10, 5, 2, 1]

def calcul_bancnote(suma_initiala):
    
    suma = suma_initiala
    response_dict = { }

    for val_banc in BANCNOTE:
        nr_banc = suma // val_banc
        suma = suma % val_banc
        response_dict[val_banc] = nr_banc

    return response_dict
```

```python
calcul_bancnote(128)
```
