```python
# Creați o lista cu prețurile unor produse introduse de utilizator. Acestor prețuri fără TVA li se vor calcula valoarea finală, cu taxa inclusă.

# În cadrul programului este hardcodată taxa la valoarea de 21%.

TVA = 21
```

```python
preturi = [ ]
print(len(preturi), preturi)
```

```python
pret = input("Introdu urmatorul pret")

while pret:
    pret = int(pret)
    preturi.append(pret)

    pret = input("Introdu urmatorul pret")
print("Preturile introduse sunt:", preturi)
```

```python
preturi_cu_TVA = []

for p in preturi:
    p_TVA = p + p * TVA / 100
    preturi_cu_TVA.append(p_TVA)

print("Preturile cu tva sunt:", preturi_cu_TVA)

```
