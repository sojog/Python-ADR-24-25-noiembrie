# Lista (mutabil) vs Tuple (imutabil)

```python
a_list = [1, 2, 3, 4]
a_tuple = (1, 2, 3, 4)
```

```python
a_list.append(10)
print(a_list)
```

```python
a_tuple.append(10)
print(a_tuple)
```

```python
a_list.pop()
a_list.pop()
print(a_list)
```

```python
a_tuple.pop()
print(a_tuple)
```

```python
a_tuple.clear()
```

```python
a_tuple = (10, 20, 30)
print(a_tuple)
a_tuple = ("alb", "negru", "rosu")
print(a_tuple)
```

```python
a = 10
b = 20
print(a, b)
```

```python
a = 10
b = 20
a, b
```

```python
a = 10
b = 20
a
```

```python
a = 10
b = 20
c = a, b
print(c, type(c))
```

```python
a = 10
b = 20
c = (a, b)
print(c, type(c))
```

```python
m = 10, 20, 30
print(m, type(m))
```

```python
m = (10, 20, 30)
print(m, type(m))
```

```python
m = [10, 20, 30]
print(m, type(m))
```

# Packing - unpacking

```python
m = 10, 20
print(m, type(m))
```

```python
m, n = 10, 20
print(m, type(m), n, type(n))
```

```python
q, w, e = 7, 8, 9
print("q=", q)
print("w=", w)
print("e=", e)
```

```python
var1 = "dimineata"
var2 = "seara"
var1, var2 = var2, var1
print("var1=", var1)
print("var2=", var2)
```
