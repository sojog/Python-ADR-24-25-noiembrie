```python
# Varianta nested (imbricat) - nepreferat
varsta = 23

if varsta < 18:
    print("Minor") 
else:
    if varsta >= 65:
        print("Pensionar")
    else:
        print("Adult")
```

```python
# Varianta flatten - preferata
varsta = 23

if varsta < 18:
    print("Minor") 
elif varsta >= 65:
    print("Pensionar")
else:
    print("Adult")
```
